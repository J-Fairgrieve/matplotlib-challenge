## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "Data/Mouse_metadata.csv"
study_results_path = "Data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, on = "Mouse ID", how = "outer")

# Display the data table for preview
combined_df
```

```python
# Checking the number of mice.
mice_count = len(combined_df["Mouse ID"].value_counts())
print(f"There are {mice_count} mice in total.")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = combined_df.loc[combined_df.duplicated(subset = ["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
print(f"The duplicate mouse is ID:{duplicates[0]}.")
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dupe_info = combined_df[combined_df["Mouse ID"] == duplicates[0]]
dupe_info
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = combined_df[combined_df["Mouse ID"].isin(duplicates) == False]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_count = len(clean_df["Mouse ID"].value_counts())
print(f"There are now {clean_count} mice in total following the data cleanse.")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

mean = clean_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
median = clean_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
variance = clean_df.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
stdev = clean_df.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
sem = clean_df.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

summary = pd.DataFrame({"Mean": mean,
                        "Median": median,
                        "Variance": variance,
                        "St. Dev": stdev,
                        "SEM": sem})

summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
agg_sum = clean_df.groupby("Drug Regimen").agg(["mean", "median", "var", "std", "sem"])["Tumor Volume (mm3)"]
agg_sum
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
capo = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:]
rami = clean_df.loc[clean_df["Drug Regimen"] == "Ramicane", :]
infu = clean_df.loc[clean_df["Drug Regimen"] == "Infubinol", :]
ceft = clean_df.loc[clean_df["Drug Regimen"] == "Ceftamin", :]

# Start by getting the last (greatest) timepoint for each mouse
capo_max = pd.DataFrame(capo.groupby('Mouse ID').max()['Timepoint'])
rami_max = pd.DataFrame(rami.groupby('Mouse ID').max()['Timepoint'])
infu_max = pd.DataFrame(infu.groupby('Mouse ID').max()['Timepoint'])
ceft_max = pd.DataFrame(ceft.groupby('Mouse ID').max()['Timepoint'])

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
capo_tumor = pd.merge(capo_max, combined_df, on = ("Mouse ID", "Timepoint"), how = "left")
rami_tumor = pd.merge(rami_max, combined_df, on = ("Mouse ID", "Timepoint"), how = "left")
infu_tumor = pd.merge(infu_max, combined_df, on = ("Mouse ID", "Timepoint"), how = "left")
ceft_tumor = pd.merge(ceft_max, combined_df, on = ("Mouse ID", "Timepoint"), how = "left")
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
box_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

# Locate the rows which contain mice on each drug and get the tumor volumes
capo_vol = capo_tumor["Tumor Volume (mm3)"]
rami_vol = rami_tumor["Tumor Volume (mm3)"]
infu_vol = infu_tumor["Tumor Volume (mm3)"]
ceft_vol = ceft_tumor["Tumor Volume (mm3)"]

box_data.append(capo_vol)
box_data.append(rami_vol)
box_data.append(infu_vol)
box_data.append(ceft_vol)
    
# add subset 
capo_quarts = capo_vol.quantile([0.25, 0.5, 0.75])
capo_lower = capo_quarts[0.25]
capo_upper = capo_quarts[0.75]
capo_iqr = capo_upper = capo_lower

rami_quarts = rami_vol.quantile([0.25, 0.5, 0.75])
rami_lower = rami_quarts[0.25]
rami_upper = rami_quarts[0.75]
rami_iqr = rami_upper = rami_lower

infu_quarts = infu_vol.quantile([0.25, 0.5, 0.75])
infu_lower = infu_quarts[0.25]
infu_upper = infu_quarts[0.75]
infu_iqr = infu_upper = infu_lower

ceft_quarts = ceft_vol.quantile([0.25, 0.5, 0.75])
ceft_lower = ceft_quarts[0.25]
ceft_upper = ceft_quarts[0.75]
ceft_iqr = ceft_upper = ceft_lower
    
# Determine outliers using upper and lower bounds
capo_lowlim = capo_lower - (1.5 * capo_iqr)
capo_upplim = capo_upper + (1.5 * capo_iqr)

rami_lowlim = rami_lower - (1.5 * rami_iqr)
rami_upplim = rami_upper + (1.5 * rami_iqr)

infu_lowlim = infu_lower - (1.5 * infu_iqr)
infu_upplim = infu_upper + (1.5 * infu_iqr)

ceft_lowlim = ceft_lower - (1.5 * ceft_iqr)
ceft_upplim = ceft_upper + (1.5 * ceft_iqr)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig, ax = plt.subplots()
ax.boxplot(box_data, labels = treatments)
ax.set_xlabel("Drug")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_title("Variation in Final Tumor Volume")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
line = capo.loc[capo["Mouse ID"] == "l509"]
x_axis = line["Timepoint"]
y_axis = line["Tumor Volume (mm3)"]

plt.plot(x_axis, y_axis)
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment over time of l509")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
scatter = capo.groupby(["Mouse ID"]).mean()
x_axis = scatter["Weight (g)"]
y_axis = scatter["Tumor Volume (mm3)"]

plt.scatter(x_axis, y_axis)
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Avg. Mouse Weight vs. Avg. Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = round(st.pearsonr(x_axis, y_axis)[0],2)
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_axis, y_axis)
regression = slope * x_axis + intercept
```

```python
print(f"The correlation between avg. mouse weight & avg. tumor volume is {correlation}")
plt.scatter(x_axis,y_axis)
plt.plot(x_axis,regression,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Avg. Mouse Weight vs. Avg. Volume")
plt.show()
```

# Observations

1. With a correlation coefficient of 0.84 it is clear that the heavier the mouse, the higher the volume of tumour it has.
2. Mice that undergo the Capomulin regimen have a lower average tumour volume, which could suggest that is the most effective drug. The effectiveness is displayed in the line graph with the tumour on subject l509 losing volume over time.
3. Out of all of the drugs displayed on the boxplot, only Infubinol has an outlier. The outlier shows a much smaller tumour volume than expected for that treatment.
